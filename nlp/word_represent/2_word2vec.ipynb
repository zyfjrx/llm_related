```python
from gensim.models import Word2Vec
import pandas as pd
import jieba
```

```python
df = pd.read_csv("./data/online_shopping_10_cats.csv",encoding="utf-8").dropna()
print(df.head())
```

```python
sentences = [[token for token in jieba.lcut(review) if token.strip() != ''] for review in df['review']]
```

```python
sentences
```

```python
model = Word2Vec(
    sentences,
    vector_size=100,
    window=5,
    min_count=2,
    sg=1,
    workers=4
)
```

```python
model.wv.save_word2vec_format('./data/word3vec.txt')
```

```python
# 加载自己训练词向量
my_model = KeyedVectors.load_word2vec_format('./data/word3vec.txt')
```

```python
my_model.most_similar(positive=['公交'])
```
