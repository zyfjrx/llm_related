```python
from cProfile import label

# 画图测试 y=sinx及其导数
import numpy as np
import matplotlib.pyplot as plt
from numpy import dtype
```

```python
# 生成数据
x = np.arange(0,6,0.1)
y = np.sin(x)
print(x,y)
```

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of sin(x)')
plt.show()
```

```python
# 画出导数 y‘ = cosx
y1 = np.cos(x)
plt.plot(x,y1)
plt.xlabel('x')
plt.ylabel('y1')
plt.title('Plot of cos(x)')
plt.show()
```

```python
plt.plot(x,y,label="y=sinx")
plt.plot(x,y1,label="y=cosx",linestyle="--")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of cos(x)')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # 导入 3D 绘图模块
```

```python
# 定义函数
def f(x, y):
    return x**2 + x*y + y**2

# 生成 x 和 y 的网格数据
x = np.linspace(-5, 5, 100)  # x 范围 [-5, 5]
y = np.linspace(-5, 5, 100)  # y 范围 [-5, 5]
X, Y = np.meshgrid(x, y)     # 生成网格点
Z = f(X, Y)                  # 计算 z = f(x, y)

# 创建 3D 图形
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# 绘制曲面
surf = ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)  # 使用渐变色

# 添加颜色条
fig.colorbar(surf, shrink=0.5, aspect=5)

# 设置坐标轴标签
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot of $f(x, y) = x^2 + xy + y^2$')

plt.show()
```

```python
# 向量范数norm
# L0范数 x=（0，2，-1）||x|| = 2
# L2范数及欧几里得距离
l0_norm = np.linalg.norm(x,ord=0)
print(l0_norm)
```

```python
l1_norm = np.linalg.norm(x,ord=1)
print(l1_norm)
```

```python
l2_norm = np.linalg.norm(x,ord=2)
print(l2_norm)
```
