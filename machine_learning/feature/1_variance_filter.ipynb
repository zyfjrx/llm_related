```python
# 特征选择 - 低方差过滤法
import numpy as np
```

```python
a = np.random.randn(100)
print(a)
print(np.var(a))
```

```python
# b = np.random.randn(100) * 0.1
b = np.random.normal(5, 0.1, 100)
print(b)
print(np.var(b))
```

```python
# 构建数据样本
X = np.vstack((a, b)).T
print(X)
print(X.shape)
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
# 低方差过滤
vt = VarianceThreshold(0.01)
X_new = vt.fit_transform(X)
print(X_new)
print(X_new.shape)
```
