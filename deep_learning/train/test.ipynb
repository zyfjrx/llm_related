```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x, y):
    return x ** 2 + x * y + y ** 2
```

```python
# 生成数据点
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# 创建3D图形
fig = plt.figure(figsize=(12, 15))
ax = fig.add_subplot(111, projection='3d')

# 绘制曲面
surf = ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)

# 添加标签和标题
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')


plt.show()

```
